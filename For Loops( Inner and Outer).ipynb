*Write a program that generates 100 random integers between 0 and 9.*

**1. Print them in a 10 by 10 matrix neatly arranged like the following (one space between each number)**

**2.If the random number is an odd number, print "@" instead. like the following**

**3.Calculate and Print the total of each row like the following (use "*" to separate the total from the numbers)**

## Solution to Question #1

```
import random # Numbers needed to be randomly generated use random 
for outer_i in range(10): # Provide the input for amount of numbers generated per row
     for inner_i in range(10): # Loops the line ten times, would be one line otherwise 
      print(random.randint(0,9),end=" ") # Print random numbers from 0-9, the end statement spaces the numbers out to where they fulfill conditions of the assignment   

     print("\n") # Changes the dimensions to where its a ten by ten matrix 
```

##Solution to Question #2

```
for outer_i in range(10): # This is similar to the code in previous solutions, modifications are made later on 
     for inner_i in range(10):
        x=random.randint(0,9) 
        if x %2==0: # This is similar to the code above however conditions need to met in order to provide a replacment of the odd numbers, an if and else statment is introduced, to seperate odds and evens  
          print (x,end=" ") # If even then print the number as is plus spacing from end="", a reminder of 0 yields an even number 
        else :
          print ("@",end=" ") # If odds replace with "@", conidtion 2
        

     print("\n")  # Dimensions are changed 
```

```
for outer_i in range(10):
    total=0 # You can change this to add to the total sum of the row if you were like, for example total=100 would add by 100 
    for inner_i in range(10):
        x=random.randint(0,9) 
        total=total+x
        print (x,end=" ")
      
        
     
    print("*",total) # This is similar to the code in previous solutions, however the key difference is this line stated here, it adds a * to the end and a total (which is 0+x) is the sum of the rows 
    
```
