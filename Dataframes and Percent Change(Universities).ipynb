```python
COLUMN_LIST = ["UNITID", "INSTNM", "STABBR", "REGION", "ADM_RATE", "TUITIONFEE_IN"] # Only columns desired 
df_list = []                                                                        # Append empty list with csv's 

for file in file_names:                                         # looping through each file 
    print("Reading file " + file + " ......")                   # Prompts statement reading file name from file path 
    tmp_df = pd.read_csv(FILE_PATH + file, usecols=COLUMN_LIST) # Grab files from path, and read each file 
    tmp_df["YEAR"] = file[6:10]                                 # Very important, need take out _17etc, and only include years in YYYY format with file [6:10]
    df_list.append(tmp_df)                                      # Append empty list 
    
df = pd.concat(df_list, ignore_index = True)                    # Combine into one data frame 
df.info()
```

```python
def calculate_percentage(tuition_list):  # Use function like before, same as before but defining a function as a list comprehension stated earlier 
    pct_change = []
    prev_year_tuition = np.NaN
    for tuition in tuition_list:
        pct_change.append((tuition - prev_year_tuition) * 100 / prev_year_tuition)
        prev_year_tuition = tuition
    pct_change = [round(x, 2) for x in pct_change]  
    return pct_change
```
